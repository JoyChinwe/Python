```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
covid = pd.read_csv (r"C:\Users\chinw\Downloads\covid.csv")
covid
```

```python
# To divide the deaths_per _M by the cases_per_M and multiply by 100.
mortality_rate = 100 * covid['mortality_rate'] / covid['cases_per_M'] 

covid = covid.assign(mortality_rate = mortality_rate)
covid['mortality_rate'].describe()

fig, ax = plt.subplots(1, 2, figsize=(50, 4))
covid['mortality_rate'].hist(bins = 47, ax = ax[0])
covid['mortality_rate'].plot.box(title = 'Percentage rate', ax = ax[1])
print("mean mortality_rate = ", covid['mortality_rate'].mean(), "skewness of mortality_rate = ", covid['mortality_rate'].skew(),)
```

The histogram chart shows that Mortality_ rate have a positively or right skewed distribution.

```python
covid.plot.scatter(x = 'mortality_rate', y='median_age')
print("Pearson Correlation between mortality_rate and rate is ", covid.mortality_rate.corr(covid['median_age']))
print("Spearman Correlation between mortality_rate and rating is ", covid.mortality_rate.corr(covid['median_age'], method = "spearman"))
```

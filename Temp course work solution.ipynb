# 7. The statistics of the cases_per_M variable of the countries from Europe over the four 3-month periods ending May-20, Aug-20, Nov-20 and Feb-21

```python
import seaborn
tempEuropeCovid = covid.loc[(covid['continent'] =='Europe')]
tempEuropeCovid
```

```python
europeCovid = tempEuropeCovid[['cases_per_M', 'location', 'date']]
europeCovid
```

```python
# Statistics of Cases_per_M in Europe.
europeCovid.describe()
```

```python
print("Using pandas module functions")
print("  n = ", europeCovid['cases_per_M'].count())
print("Measures of central location")
print("Mean = ", europeCovid['cases_per_M'].mean())
print("Median = ", europeCovid['cases_per_M'].median())

print("Measures of spread")
print("range = ", europeCovid['cases_per_M'].max()-europeCovid['cases_per_M'].min())
q = europeCovid['cases_per_M'].quantile([0.25,0.5,0.75])
print("quartiles = ", q)
print("IQR = ", europeCovid['cases_per_M'].quantile(0.75) - europeCovid['cases_per_M'].quantile(0.25))
print("stdev = ", europeCovid['cases_per_M'].std())
print("pop stdev = ", europeCovid['cases_per_M'].std(ddof = 0))

# print("Measures of shape")
print("Skewness = ", europeCovid['cases_per_M'].skew())
print("Excess kurtosis = ", europeCovid['cases_per_M'].kurt())
plt.show()
```

```python
ct2 = pd.crosstab(europeCovid['location'],europeCovid['date'], values = europeCovid['cases_per_M'], aggfunc = 'mean')
fig, ax = plt.subplots(1, 1, figsize=(30, 8))
europeCovid['date']= pd.Categorical(europeCovid['date'], categories=['May-20','Aug-20','Nov-20','Feb-21'])
ct2.plot.bar(stacked=True, ax=ax)
plt.show()
import warnings
warnings.simplefilter('ignore')
```

The distribution shows that cases in Europe was the highest in the 3 months period ending in Feb 2021, which started in the 3 month period ending in Nov-20. This means that cases were at the highest during the winter period in Europian countries.
Czechoslovakia has the highest cases in Europe for the 3 months period ending in Nov-20 and Feb-21 followed by Slovenia for the same 3 months periods.
The graph above show that Finland has the least cases for the 3 months periods(May-20,Aug-20,Nov-20,Feb-21).
